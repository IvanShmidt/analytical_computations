```python
# shape >= (2, 2), degree of 2
def W_matr(shape):
    if shape != (2,2):
        limit = shape[0]//2
        a = W_matr((limit, limit))
        b = np.zeros([2*i for i in a.shape])
        for i in range(a.shape[0]):
            for j in range(a.shape[0]):
                b[i,j]             = a[i,j]
                b[i+limit,j]       = a[i,j]
                b[i,j+limit]       = a[i,j]
                b[i+limit,j+limit] = -1*a[i,j]
        return b
    else:
        return np.array([[1, 1], [1, -1]])
    
    
def funk_vector(funk, dim):
    x = []
    for i in range(2 ** dim):
        #vec = list('0'*(7-len(bin(i)[2:])) + bin(i)[2:])
        vec = list('0'*(dim-len(bin(i)[2:])) + bin(i)[2:])
        vec = np.array([int(j) for j in vec])
        #print(vec)
        x.append(funk(vec))        
        #x.append([i for i in list('0'*(7-len(bin(i)[2:])) + bin(i)[2:])])
    return(np.array(x))


def Fourier_coef(funks, dim):
    #print('Fourier coefficients for {}'.format([f.__name__ for f in funks]))
    W = W_matr((2**dim, 2**dim))
    FC = {}
    for funk in funks:
        #print('{}: {}'. format(funk.__name__, W.dot(funk_vector(funk, dim))))
        FC[funk.__name__] = W.dot(funk_vector(funk, dim))
    return FC
        
        
def exp_f(funk, dim):
    return np.array([1]*2**dim) - 2 * funk_vector(funk, dim)
    

def Walsh_coef(funks, dim):
    W = W_matr((2**dim, 2**dim))
    WC = {}
    for funk in funks:
        WC[funk.__name__] = W.dot(exp_f(funk, dim))
    return WC
        
```

```python

# Действительный многочлен

def real_polynomial_matr(shape):
    if shape != (2,2):
        limit = shape[0]//2
        a = real_polynomial_matr((limit, limit))
        b = np.zeros([2*i for i in a.shape])
        for i in range(a.shape[0]):
            for j in range(a.shape[0]):
                b[i,j]             = a[i,j]
                b[i+limit,j]       = a[i,j]*-1
                b[i,j+limit]       = a[i,j]*0
                b[i+limit,j+limit] = a[i,j]
        return b
    else:
        return np.array([[1, 0], [-1, 1]])

def Real_polynomial_coef(funks, dim):
    #print('Fourier coefficients for {}'.format([f.__name__ for f in funks]))
    W = real_polynomial_matr((2**dim, 2**dim))
    FC = {}
    for funk in funks:
        #print('{}: {}'. format(funk.__name__, W.dot(funk_vector(funk, dim))))
        FC[funk.__name__] = W.dot(funk_vector(funk, dim))
    return FC
```

```python

# вес линейных комбинаций разрядных функций

# возвращ словарь сo строкой лин комб и ее соотв вектором.
def lin_comb_funk_vec(dim):
    linear_comb = {}
    for k in range(1, 2**(dim+1)):
        vec = list(map(lambda x: int(x), '0'*(dim+1-len(bin(k)[2:])) + bin(k)[2:]))
        f_name = '⊕'.join([j*'y{}'.format(i) for i,j in enumerate(vec[1:]) if j>0])
        funk = f_name if vec[0]==0 else '1⊕'+f_name
        linear_comb[funk] = vec
    return linear_comb

#comb_funks = lin_comb_funk_vec(9)

# для вектора 
def calculate_lin_comb_funk(vec, funks, funk_vec):
    val = []
    for i, el in enumerate(funk_vec[1:]):
        val.append(el*funks[i](vec))
    val.append(funk_vec[0])
    return mod2(val)

def weight_lin_comb_funks(y_funks, dim):
    comb_funks = lin_comb_funk_vec(dim)
    weights = {}
    for f in comb_funks.keys():
        weights[f] = 0
        funk_vec = comb_funks[f]
        for i in range(2**dim):
            x = list('0'*(dim-len(bin(i)[2:])) + bin(i)[2:])
            x = np.array([int(i) for i in x])
            if calculate_lin_comb_funk(x, y_funks, funk_vec) == 1:
                weights[f] += 1
        #print(f, weights[f])
    return weights

def funk_vector_lin_comb(funk_vec, funks, dim):
    x = []
    for i in range(2 ** dim):
        #vec = list('0'*(7-len(bin(i)[2:])) + bin(i)[2:])
        vec = list('0'*(dim-len(bin(i)[2:])) + bin(i)[2:])
        vec = np.array([int(j) for j in vec])
        #print(vec)
        x.append(calculate_lin_comb_funk(vec, funks, funk_vec))      
        #x.append([i for i in list('0'*(7-len(bin(i)[2:])) + bin(i)[2:])])
    return(np.array(x))

def Fourier_coef_lin_comb(funks, dim):
    W = W_matr((2**dim, 2**dim))
    FC = lin_comb_funk_vec(dim)
    for funk_vec in FC.keys():    
        FC[funk_vec] = W.dot(funk_vector_lin_comb(FC[funk_vec], funks, dim))
    return FC

def exp_f_lin_comb(funk_vec, funks, dim):
    return np.array([1]*2**dim) - 2 * funk_vector_lin_comb(funk_vec, funks, dim)
    

def Walsh_coef_lin_comb(funks, dim):
    print('Walsh coefficient for {}'.format([f.__name__ for f in funks]))
    W = W_matr((2**dim, 2**dim))
    WC = lin_comb_funk_vec(dim)
    for funk_vec in WC.keys():    
        WC[funk_vec] = W.dot(exp_f_lin_comb(WC[funk_vec], funks, dim))
    return WC
    
    
    for funk in funks:
        print('{}: {}'. format(funk.__name__, W.dot(exp_f(funk, dim))))

# Веса лин комб
# w = weight_lin_comb_funks(s9_funks, 9)
# for k in w.keys():
#     print('{:25} {}'.format(k, w[k]))


# Коэффициенты Фурье лин комб
#fc = Fourier_coef_lin_comb(s9_funks, 9)
# for k in fc.keys():
#     print(k, 0.5 - fc[k]/2**9)

# Коэф Уолша - Адамара
wc = Walsh_coef_lin_comb(s9_funks, 9)
for k in wc.keys():
    print(k , wc[k])
```
